```python
def accept_reject(N):
    xmin = 0
    xmax = np.pi
#5
    pmax = 0.637

    n_accept=0
    x_list = [] 
    while n_accept < N:
        t = (xmax-xmin)*np.random.rand() + xmin
        y = np.random.rand()
        if y < f(t)/ pmax:
            n_accept += 1
            x_list.append(t)
    return x_list

x = accept_reject(10000)
x
```
